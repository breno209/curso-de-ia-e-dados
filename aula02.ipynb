Sistema que irá verificar se o filme avaliado é bom ou não.
-criar uma variável que irá armazenar o nome do filme
-criar uma variável que irá armanezar a nota do filme (0 - 5)
-condição: se possuir 5 estrelas, filme vencedor Oscar.
-se possuir de 3 a 4,99, filme bom para assistir
-abaixo disso, filme não indicado.

```python
gente_grande = input("digite o nome do filme")
nota_gente_grande = int(input("digite a nota do filme: "))
if nota_gente_grande == 5:
    print(f"filme (gente_grande) vencedor do Oscar")
elif nota_gente_grande >= 3:
    print(f"filme (gente_grande) bom pra assistir")
else:
    ("não indicado")
```

```python
gente_grande = input("digite o nome do filme")
nota_gente_grande = float(input("digite a nota do filme:"))
if nota_gente_grande <= 0 or nota_gente_grande > 5 :
    print("❌ nota inválida")
else:
    if nota_gente_grande == 5:
    print(f"filme (gente_grande) vencedor do Oscar")
    elif nota_gente_grande >= 3:
    print(f"filme (gente_grande) bom pra assistir")
    else:
    ("não indicado")
```
