```python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]
nomes_produtos = ["pastel","caçulinha","refrigerante","combos"]

array_produtos = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
for nome, preco in zip(nomes_produtos, faturamento):
    print(f"{nome} : {preco}")
```

### Exercício de fixação

- criar uma lista qu irá possuir 10 nomes de produtos.
- criar uma lista que irá possuir o preço de custo(atacado).
- criar uma lista que irá possuir o preço de venda (varejo).
- ao final, exibir o lucro obtido em cada produto.

```python
precos_varejo = [23.90, 7.50, 7.90, 4.90, 5.90, 16.90, 5.20, 4.50, 2.50, 2.90 ]
precos_atacado = [18.50, 6, 6.40, 3.90, 4.80, 13.80, 4.20, 3.40, 1.90, 2.20]
nomes_produtos = ["arroz", "feijão","óleo de soja", "açúcar", "farinha de trigo", "café", "leite integral", "macarraão espaguete", "molho de tomate", "sal"]

array_varejo = np.array(precos_varejo)
array_atacado = np.array(precos_atacado)

faturamento = array_varejo - array_atacado
for nome, preco in zip(nomes_produtos, faturamento):
    print(f"o lucro do produto {nome} foi {preco:.2f}")
    
    
```
