## *Separando os tópicos que são númericos*

```python
colunas = [
    "age",
    "study_hours_per_day",
    "social_media_hours",
    "netflix_hours",
    "attendance_percentage",
    "sleep_hours",
    "exercise_frequency",
    "mental_health_rating",
    "exam_score"
]
```

## *Gráfico de análise dos tópicos númericos*

```python
sns.heatmap(data=df[colunas].corr(), annot=True, cmap="coolwarm",fmt=".2f");
```

## *Gráfico da base de dados das notas e quantidade de horas estudadas por dia das pessoas*

```python
sns.lmplot(data=df,x="study_hours_per_day",y="exam_score");
```

## *Gráfico da quantidade de horas que as pessoas ficaram asisitindo netflix e suas notas*

```python
sns.lmplot(data=df,x="netflix_hours",y="exam_score");
```

## *Cálculo da média das pessoas correspondente ao tanto de horas estudadas por dia*

```python
filtro_horas_altas = df["study_hours_per_day"] >= 5
filtro_horas_baixas = df["study_hours_per_day"] < 2

media_alta = df[filtro_horas_altas]["exam_score"].mean()
media_baixa = df[filtro_horas_baixas]["exam_score"].mean()

print(f"a média das pessoas que estudaram 5 ou mais horas foi {media_alta:.2f}. já a média, das pessoas que estudaram menos de duas horas, foi {media_baixa:.2f}")
```
