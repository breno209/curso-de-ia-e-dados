## *Cálculo da média dos estudantes correspondente ao tanto de horas estudadas por dia*

```python
filtro_horas_altas = df["study_hours_per_day"] >= 5
filtro_horas_baixas = df["study_hours_per_day"] < 2

media_alta = df[filtro_horas_altas]["exam_score"].mean()
media_baixa = df[filtro_horas_baixas]["exam_score"].mean()

print(f"a média das pessoas que estudaram 5 ou mais horas foi {media_alta:.2f}. já a média, das pessoas que estudaram menos de duas horas, foi {media_baixa:.2f}")
```

## *Pergunta 2, Esta base tem 1000 alunos, com hábitos (horas de estudo, sono, redes sociais, exercício) e a nota final no exame (exam_score). Analise quais hábitos mais influenciam a nota e mostre os números que comprovam.*

## *Como analisei Fiz duas análises complementares: 1. Correlação simples (cada hábito isoladamente com a nota) 2. Regressão múltipla padronizada (todos os 7 hábitos juntos, controlando uns pelos outros — isso mostra o efeito "líquido" de cada um) O modelo de regressão múltipla explica 90,1% da variação nas notas (R² = 0,901), um ajuste muito forte.

---

## *Conclusão - Horas de estudo por dia é, disparado, o hábito mais determinante. Sozinho, explica 68% da variação nas notas (r² = 0,681). No modelo completo, controlando por sono, redes sociais, exercício, presença e saúde mental, um aumento de 1 desvio-padrão nas horas de estudo (~2,3h/dia) ainda está associado a +14 pontos na nota — quase 3x o efeito do segundo colocado, saúde mental (+5,6 pontos). Os hábitos "de tela" (netflix e redes sociais) puxam a nota para baixo de forma consistente, mas o efeito é modesto perto do estudo. Sono e exercício ajudam moderadamente, e presença tem o menor impacto de todos os hábitos medidos.*

## *existe uma quantidade ideal de horas de sono para tirar as melhores notas?*

## *O que os gráficos revelam, O gráfico de dispersão com a curva quadrática mostra claramente o padrão em "U invertido": Dormir pouco (< 6h) prejudica visivelmente — quem dorme menos de 4h tem nota média de 60,3, quase 13 pontos abaixo do pico. O pico de desempenho está entre 7 e 9 horas — a faixa 7–8h apresenta a maior média (73,0) e mediana (73,5). Dormir demais (> 9h) não traz ganho adicional — a nota volta a cair ou se estabiliza, possivelmente porque excesso de sono reflete outros problemas (depressão, sedentarismo, etc.).*

## *Conclusão. Sim, existe uma faixa ótima de sono associada às melhores notas: entre 7 e 9 horas por noite, com pico estimado em torno de 9,3 horas. Dormir menos de 6h reduz significativamente o desempenho, enquanto dormir mais de 9h não oferece vantagem adicional.*
