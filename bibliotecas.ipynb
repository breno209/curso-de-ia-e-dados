```python
#importação de bibliotecas
import numpy as np #np será o apelido
```

```python
temperaturas = np.array([30,40,50,60,70,90,100])
print(temperaturas)
print(type(temperaturas))
```

```python
#média aritmética
media = np.mean(temperaturas)
print(media)
```

```python
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
#salários x 1000
salarios = np.array([2,3,4,5,1,500])
media_salarios = np.mean(salarios)
print(media_salarios)
print(f"{media:.2f}")
```

```python
mediana = np.median(salarios)
print(mediana)
```
